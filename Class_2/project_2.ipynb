You are a programmar, now solve the following problem with the help of python.

Chaand Taara Khan is hosting a rabri (desert) party at his institute. He needs your help to calculate the **total number os people attending** the event and the **amount of Rabri** (in grams and kg) he needs to prepare.

- There are 12 faculty members.
- There are 4 administrative staff members.
- There are 100 students in the institute.
- There are 15 people absent.
- Per person rabri is 250g

```python
num_of_faculty = 12
num_of_admins = 4
num_of_students = 100
no_of_absenties = 15
per_person_rabri_gram = 250

total_members = num_of_faculty + num_of_admins + num_of_students

total_people_attend = total_members - no_of_absenties

total_rabri_gram = total_people_attend * per_person_rabri_gram

total_rabri_kilogram = total_rabri_gram // 1000

print(f"Total Memebers Invited: {total_members}")
print(f"Total Members attended the party (present): {total_people_attend}")
print(f"Total Rabri in gram to make: {total_rabri_gram} g")
print(f"Total Rabri in kilogram to make: {total_rabri_kilogram} kg")
```
