# New Year Countdown (10 to 1)

## Bad Program

The following code is in the raw form. This is a bad practice as it is repeating itself many times.

```python
countdown : int = 10

print(f"{countdown}...")

countdown -= 1

print(f"{countdown}...")

countdown -= 1

print(f"{countdown}...")

countdown -= 1

print(f"{countdown}...")

countdown -= 1

print(f"{countdown}...")

countdown -= 1

print(f"{countdown}...")

countdown -= 1

print(f"{countdown}...")

countdown -= 1

print(f"{countdown}...")

countdown -= 1

print(f"{countdown}...")

countdown -= 1

print(f"{countdown}...")

countdown -= 1

print("Happy New Year!")
```

```python
countdown : int = 10

while countdown >= 1:
    print(f"{countdown}...")
    countdown -= 1

print("Happy New Year!")
```
